```python
import pandas as pd
import numpy as np
```

```python
df_countries = pd.read_csv('covid19.csv')
```

```python
df_countries = df_countries[1:]
```

```python
df_countries = df_countries.nlargest(15,'Cases - cumulative total')
```

```python
df_countries.to_csv('top15countries.csv',index=False)
```

```python
df_countries["count"] = 0
```

```python
df_countries
```

```python
import json
lis_codes= []
lis_flags = []
for index, row in df_countries.iterrows():
    with open('flags.json',encoding='utf-8') as data_file:    
        data = json.load(data_file)
        for v in data:
            
            if(v["name"] == row["Name"] ):
                lis_flags.append(v["flag"])
                lis_codes.append(v["code"])
                break;
df_countries["code"] = lis_codes
df_countries["flags"] = lis_flags
```
